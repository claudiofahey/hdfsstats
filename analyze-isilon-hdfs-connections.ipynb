# Analyze Isilon HDFS Connections

This IPython Notebook demonstrates how to read the log files created by hdfsstats.sh, which records HDFS connection counts on an Isilon cluster.

Required Python libraries:
-  Pandas
-  Numpy

## Imports and Function Definitions

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pylab as plt
import itertools
import datetime
import re
import glob
import re
import dateutil
```

```python
%matplotlib inline
```

```python
def text_files_iterator(src, verbose=False):
    """Iterate over a set of text files identified with a wildcard (e.g. *.log)."""
    for filename in glob.glob(src):
        if verbose: print('Reading %s ' % filename)
        with open(filename, 'r') as file:
            for line in file:
                yield line
```

```python
def read_hdfsstats(src):
    """Read files created by hdfsstats.sh and return a dataframe."""
    node_connection_count_re = re.compile('.*-([0-9]+):\\W*([0-9]+) NN, ([0-9]+) DN')  # Isi-hadoop-13: 23 NN, 6 DN
    data = []
    for line in text_files_iterator(src):
        match = node_connection_count_re.search(line)
        if match:
            data.append({'utc': timestamp, 
                         'node_number': int(match.groups()[0]),
                         'nn_connection_count': int(match.groups()[1]),
                         'dn_connection_count': int(match.groups()[2])})
        else:
            # All other lines should be a UTC timestamp
            timestamp = np.datetime64(dateutil.parser.parse(line))
    df = pd.DataFrame(data).groupby(by=['utc','node_number']).first()  # discard duplicate records
    df['connection_count'] = df.nn_connection_count + df.dn_connection_count
    return df
```

```python
def plot_hdfsstats(hdfsstats_df, utc_begin=None, utc_end=None, y_col='dn_connection_count'):
    """Plot Isilon HDFS connection counts to each Isilon node vs. time."""
    # Filter connection statistics for time period
    mask = np.ones(len(hdfsstats_df)).astype(bool)
    if utc_begin:
        mask = mask & (np.datetime64(utc_begin) <= hdfsstats_df.index.get_level_values(0))
    if utc_end:
        mask = mask & (hdfsstats_df.index.get_level_values(0) <= np.datetime64(utc_end))
    filtered_hdfsstats_df = hdfsstats_df[mask].reset_index()
    
    # Standard plot
    if True:
        line_cycler = itertools.cycle(['-','--','-.',':'])
        color_cycler = itertools.cycle(['r','b','g','c','m','y','k'])
        fig = plt.figure(figsize=(20,10))
        fig.clf()
        axes = plt.gca()
        for name, group in filtered_hdfsstats_df.groupby(['node_number']):
            axes.plot(group.utc.astype(datetime.datetime).values, 
                      group[y_col].values, 
                      next(color_cycler) + next(line_cycler),
                      label='Node %d' % name)
        fig.autofmt_xdate()
        axes.set_xlabel('UTC')
        axes.set_ylabel(y_col)
        axes.legend(loc='best')        
        plt.show()            
        
    # Stacked plot
    if True:
        # Transform data for plotting
        plot_df = filtered_hdfsstats_df.pivot(index='utc', columns='node_number', values=y_col)        
        # Order nodes by mean connection count so the busiest ones are together at the top
        node_connection_count_mean = filtered_hdfsstats_df.groupby(by=['node_number'])[y_col].mean()
        node_connection_count_mean.sort()
        node_list = node_connection_count_mean.index.values
        plot_df = plot_df[node_list]
        # Create stack plot
        fig = plt.figure(figsize=(20,10))
        fig.clf()
        axes = plt.gca()
        cmap=plt.get_cmap('Set3')
        colors = cmap(np.linspace(0, 1, len(plot_df.columns)))
        axes.stackplot(plot_df.index.astype(datetime.datetime).values, plot_df.values.T, colors=colors)
        axes.set_xlabel('UTC')
        axes.set_ylabel(y_col)
        plt.show()
```
